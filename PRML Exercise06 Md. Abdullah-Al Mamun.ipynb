```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
from sklearn.ensemble import*


# loading dataset
cancer_dataset=loadmat("arcene.mat")

trainX=cancer_dataset['X_train']
trainy=cancer_dataset['y_train'].ravel()
testX=cancer_dataset['X_test']
testy=cancer_dataset['y_test'].ravel()

# traing a random forest classifier with 100 trees.

classifier=RandomForestClassifier(100)
classifier.fit(trainX,trainy)

# plotting a histogram of its feature importance.

plt.figure(figsize=(10,6),facecolor=".9")

a=list(range(trainX.shape[1]))
b=classifier.feature_importances_

plt.bar(a,b,color="r", width=3)
plt.ylim([0, classifier.feature_importances_.max()])
plt.title("Histogram of its feature importance")
plt.xlabel("Important Features")
plt.ylabel("Coefficients")

ax = plt.gca()
ax.set_facecolor('#F5E1A4')


plt.show()

# Accuracy of X_test and y_test

accuracy=classifier.score(testX,testy)
print("Accuracy :{}".format(accuracy))
```

```python
import matplotlib.pyplot as plt
from sklearn.feature_selection import*
import numpy as np
from sklearn.discriminant_analysis import*


linear_discriminat_analysis=LinearDiscriminantAnalysis()
rfecv=RFECV(linear_discriminat_analysis,verbose=1, step=50)
rfecv.fit(trainX,trainy)
```

```python

features=rfecv.support_.sum()
print("Number of selected features: {}".format(features))


# Plotting
plt.figure(figsize=(10,6),facecolor=".9")
plt.plot(range(0,10001,50), rfecv.grid_scores_, c='red')
plt.xlabel("Features")
plt.ylabel("Error Rate")
plt.title("Errors for different Number of Features")
plt.grid()

ax = plt.gca()
ax.set_facecolor('#6AD1E3')
plt.show()


accuracy=rfecv.score(testX,testy)
print("Accuracy: {}".format(accuracy))
```

```python
# importing libraries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import*
from sklearn.linear_model import*


# using Logistic Regression classifier

logistic_regression=LogisticRegression(solver='liblinear', penalty='l1', verbose=0)

# Finding the accuracy of the classifier with C.

bestC=[10**i for i in range(-4,3)]
grid_search=GridSearchCV(logistic_regression,{'C':bestC},cv=5)
grid_search.fit(trainX,trainy)
dataframe_result=pd.DataFrame(grid_search.cv_results_)

best_parameter=grid_search.best_params_
print("The best parameter C: {}".format(best_parameter))

# Fitting the Logistic Regression model to X_train and y_train with the best C.

best_logistic_regression=LogisticRegression(solver='liblinear', penalty='l1', C=grid_search.best_params_['C'])
best_logistic_regression.fit(trainX,trainy)

# Counting the number of selected features
best_featurecount=best_logistic_regression.coef_.nonzero()[1].shape[0]
print("The number of selected features: {}".format(best_featurecount))

# Finding the accuracy on X_test and y_test 
accuracy=best_logistic_regression.score(testX,testy)
print("The accuracy on X_test and y_test: {}".format(accuracy))
```
